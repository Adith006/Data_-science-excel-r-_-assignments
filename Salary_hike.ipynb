# Building model without transformations

```python
x = dataset.YE
y = dataset.Sal
```

```python
#building our first model
model_1 = sm.ols("Sal~YE",data = dataset).fit()
```

```python
#plot of best line in the no transformation case
plt.plot(x,model_1.params[0]+model_1.params[1]*(x),"r")
sn.scatterplot(x,y)
```

```python
#cumulating the model summary
model_1.summary()
```

# since our accuracy is 95.6% may be our model is overfitting 

# Log transformation with x and y

```python
log_x = np.log(x)
log_y = np.log(y)
dataset["log_x"]= log_x
dataset["log_y"] = log_y
```

```python
#plotting with original data and log data
plt.subplots(1,1)
plt.suptitle("years experienced")
sn.distplot(x)
sn.distplot(log_x)
labels = ("original data","transformed data")
plt.legend(labels)

plt.subplots(1,1)
plt.suptitle("salary")
sn.distplot(y)
sn.distplot(log_y)

labels = ("original","Transformed data")
plt.legend(labels)
plt.show()
```

# with log transformed X and Y building model

```python
model_2 = sm.ols("log_y~log_x",data = dataset).fit()
```

```python
#plot of best line in the log transformation case
plt.plot(log_x,model_2.params[0]+model_2.params[1]*(log_x),"r")
sn.scatterplot(log_x,log_y)
```

```python
model_2.summary()
```

# After log transformed the accuracy is 90 %


# Square root Transformation

```python
sqrt_x = np.sqrt(x)
sqrt_y = np.sqrt(y)
dataset["sqrt_x"]= sqrt_x
dataset["sqrt_y"] = sqrt_y
```

```python
#plotting with original data and log data
plt.subplots(1,1)
plt.suptitle("years experienced")
sn.distplot(x)
sn.distplot(sqrt_x)
labels = ("original data","transformed data")
plt.legend(labels)

plt.subplots(1,1)
plt.suptitle("salary")
sn.distplot(y)
sn.distplot(sqrt_y)

labels = ("original","Transformed data")
plt.legend(labels)
plt.show()
```

# with square root of X and Y building Model

```python
model_3 = sm.ols("sqrt_y~sqrt_x",data = dataset).fit()
```

```python
#plot of best line in the log transformation case
plt.plot(sqrt_x,model_3.params[0]+model_3.params[1]*(sqrt_x),"r")
sn.scatterplot(sqrt_x,sqrt_y)
```

```python
model_3.summary()
```

# with square root Transformation the accuracy is 94 %

# checking with square Transformation

```python
sqr_x = np.square(x)
dataset["sqr_x"] = sqr_x
sqr_y = np.square(y)
dataset["sqr_y"] = sqr_y
```

```python
#plotting with original data and log data
plt.subplots(1,1)
plt.suptitle("years experienced")
sn.distplot(x)
sn.distplot(sqr_x)
labels = ("original data","transformed data")
plt.legend(labels)

plt.subplots(1,1)
plt.suptitle("salary")
sn.distplot(y)
sn.distplot(sqr_y)

labels = ("original","Transformed data")
plt.legend(labels)
plt.show()
```

```python
#building model for square transformation
model_4 = sm.ols("sqr_y~sqr_x",data = dataset).fit()
```

```python
#plot of best line in the square transformation case
plt.plot(sqr_x,model_4.params[0]+model_4.params[1]*(sqr_x),"r")
sn.scatterplot(sqr_x,sqr_y)
```

```python
model_4.summary()
```

# The accuracy with square tranformation is 95.4%

# after all analysis we conclude that  our best accuracy is without transformation
